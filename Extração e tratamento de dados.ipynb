## Coleta informações dos artistas

```python
def clean_response_artist(api_response):
    """
    Filtra algums informações importantes para o negócio, dentre todas as outras disponíveis na resposta da API.
    
    followers_of_artist -> Quantidade de seguidores que um artista tem na plataforma.
    artist_id -> ID único do artista dentro da plataforma
    artist_name -> Nome oficial do artista/banda
    artist_popularity -> Popularidade do artista na plataforma.
                         É um valor de 0 a 100, sendo que 100 corresponde ao maior nível de popularidade possível.
                         A popularidade de um artista é calculada a partir da popularidade de todas as faixas do artista.
    
    Referência: https://developer.spotify.com/documentation/web-api/reference/#object-artistobject
    """
    followers_of_artist = api_response[0]['followers']['total']
    artist_id = api_response[0]['id']
    artist_name = api_response[0]['name']
    artist_popularity = api_response[0]['popularity']
    
    meta_artist = {'followers': followers_of_artist,
                  'artist_id': artist_id,
                  'artist_name': artist_name,
                  'popularity': artist_popularity}
    
    return meta_artist


def create_artist_table(artist_name, genre):
    sr = sp.search(q=artist_name, type='artist', limit=1)['artists']['items']
    df_artist = pd.DataFrame.from_dict(clean_response_artist(sr), orient='index').transpose().set_index('artist_name')
    df_artist['genre'] = genre
    
    return df_artist
```

```python
genres = {'Rap': artistas_rap,
         'Rock': artistas_rock,
          'Sertanejo': artistas_sertanejo,
          'Axe': artistas_axe,
          'Pagode': artistas_pagode}
```

```python
all_artists = []

for estilo, artistas in genres.items():
    for artista in artistas:
        all_artists.append(create_artist_table(artist_name=artista, genre=estilo))
```

```python
all_artists_df = pd.concat(all_artists)
all_artists_df['popularity'] = all_artists_df['popularity'].astype(int)
```

## Artistas relacionados

```python
network_path = Path('network_graph')
if not network_path.exists():
    network_path.mkdir()
```

```python
def create_artists_relations(artist_seed, artist_name):
    """
    Busca na API do Spotify quais artistas são relacionados com o que foi usado como input.
    artist_seed deve ser o ID do artista na plataforma.
    Retorna um pd.DataFrame com os artistas relacionados dispostos na coluna 'target'.
    """
    related = pd.DataFrame.from_dict(sp.artist_related_artists(artist_seed)['artists'])
    followers_of_related = pd.json_normalize(related.followers).total
    related = pd.concat([related, followers_of_related], axis=1)
    related['source'] = artist_name
    related.drop(columns=['external_urls', 'followers', 'genres', 'href', 'images', 'uri', 'type'], inplace=True)
    related.rename(columns={'name': 'target',
                           'total': 'target_followers',
                           'popularity': 'target_popularity',
                           'id': 'target_id'}, inplace=True)
    related.set_index('target', inplace=True)
    return related
    
```

```python
df_related = []
for index, row in all_artists_df.iterrows():
    relations = create_artists_relations(artist_seed=row.artist_id, artist_name=index)
    df_related.append(relations)
```

```python
graph_network = pd.concat(df_related)
graph_network['target_popularity'] = graph_network['target_popularity'].astype(int)
graph_network['target_followers'] = graph_network['target_followers'].astype(int)
```

```python
graph_network.to_csv("network_graph/network_graph_between_artists.csv")
graph_network
```

```python
def create_top_tracks_table(artist_id):
    df_music = pd.DataFrame.from_dict(sp.artist_top_tracks(artist_id, country='BR')['tracks'])
    album_details = pd.json_normalize(df_music.album)
    music_artists = pd.json_normalize(album_details.artists.explode())

    df_music.drop(columns=['uri', 'type', 'track_number', 'preview_url', 'href', 'external_urls', 'external_ids', 'disc_number', 'artists', 'album', 'is_local'], inplace=True)
    album_details.drop(columns=['external_urls.spotify', 'uri', 'type', 'images', 'href', 'artists', 'album_type'], inplace=True)
    music_artists.drop(columns=['href', 'type', 'uri', 'external_urls.spotify'], inplace=True)

    for c in df_music.columns:
        df_music.rename(columns={c: f"top_track_{c}"}, inplace=True)

    for c in album_details.columns:
        album_details.rename(columns={c: f"album_{c}"}, inplace=True)

    for c in music_artists.columns:
        music_artists.rename(columns={c: f"artist_{c}"}, inplace=True)  

    top_track_single_artist = pd.concat([df_music, album_details, music_artists], axis=1)
    
    return top_track_single_artist
    
```

## Top tracks features

```python
track_features_path = Path('track_features')
if not track_features_path.exists():
    track_features_path.mkdir()
```

```python
def clean_track_features(track_id):
    """
    Recupera informações a respeito das características da faixa de acordo com as análises produzidas pelo Spotify.
    Retorna um pd.DataFrame com as seguintes métricas específicas:
    
    'danceability' -> Demonstra quanto uma faixa é dançável de acordo com elementos contidos nela
                    como ritmo, força das batidas, etc...
                    Um valor 0.0 representa uma faixa pouco dançável, e um valor de 1.0 uma faixa muito dançável.
    'energy' -> Representa uma percepção de intensidade da faixa. Música clássica tende a ter uma intensidade baixa,
                portanto recebe valores de energy menores (~0.0). Músicas de festa recebem notas maiores (~1.0)
    'key', -> Nota musical que rege a música. Corresponde a seguinte notação:
            0: C: Dó
            1: C#
            2: D: Ré
            3: D#
            4: E: Mi
            5: F: Fa
            6: F# 
            7: G: Sol
            8: G#
            9: A: La
            10: A#
            11: B: Si
    'loudness'-> Média da intensidade da faixa, medida em decibels(dB). 
    'mode', -> Indica o tipo de escala em que a melodia é construída (maior/menor). Maior == 1; Menor == 0.
    'speechiness', -> Detecta a existência de vocal em uma faixa.
                    Valores acima de 0.66 geralmente são dados à faixas feitas exclusivamente de vocalizações (podcasts, p.ex.)
                    Valores entre 0.33 e 0.66 são declarados para faixas que contém música e vocal, separados ou sobrepostos.
                    Valores abaixo de 0.33 são declarados para faixas que contém muito pouca ou nenhuma participação de vocal.
    'acousticness', -> Valor de confidência que indica se uma faixa é acústica ou não.
                    Valores próximos a 1.0 indicam alta confiança de que a faixa é acústica
    'instrumentalness', -> Indica se uma faixa contém vocais.
                        Faixas sem vocais recebem valores mais perto de 1.0, ou seja, são compostas majoritariamente de instrumentos.
    'liveness', -> Detecta presença de audiência participante na gravação.
                    Valores perto de 1.0 indicam que a faixa foi possivelmente grava ao vivo
    'valence', -> Indica o humor da faixa.
                Valores mais próximos de 1.0 são reservados para faixas com 'energia positiva'.
                Valores mais próximos de 0.0 indicam faixas mais tristes ou raivosas.
    'tempo', -> Ritmo da faixa, medido em BPM (batidas por minuto).
    'time_signature' -> Estimativa de quantas batidas há em uma medida.
    
    """
    
    tf = sp.audio_features(track_id)
    single_tf = pd.DataFrame.from_dict(tf)
    single_tf.drop(columns=['type', 'uri', 'track_href', 'analysis_url'], inplace=True)
    
    single_tf['key'] = single_tf['key'].astype(int)
    single_tf['mode'] = single_tf['mode'].astype(int)
    single_tf['instrumentalness'] = single_tf['instrumentalness'].astype(int)
    single_tf['duration_ms'] = single_tf['duration_ms'].astype(int)
    single_tf['time_signature'] = single_tf['time_signature'].astype(int)
    
    return single_tf
```

```python
tt_features = []
for index, row in top_tracks_df.iterrows():
    track_feature = clean_track_features(track_id=index)
    tt_features.append(track_feature)
```

```python
top_track_features = pd.concat(tt_features)
top_track_features.drop_duplicates('id', inplace=True)
top_track_features.dropna(inplace=True)
top_track_features.set_index('id', inplace=True)
```

```python
def create_album_table(artist_id):
    album_response = pd.DataFrame.from_dict(sp.artist_albums(artist_id, country='BR', album_type='album')['items'])
    album_artists = pd.json_normalize(album_response.artists.explode())

    for c in album_response.columns:
        if 'album' not in c:
            album_response.rename(columns={c: f"album_{c}"}, inplace=True)

    for c in album_artists.columns:
        album_artists.rename(columns={c: f"artist_{c}"}, inplace=True)

    album = pd.concat([album_response, album_artists], axis=1)
    album.drop(columns=['album_group',
                        'album_type',
                        'album_artists',
                        'album_external_urls',
                        'album_href',
                        'album_images',
                        'artist_href',
                        'artist_type',
                        'artist_uri',
                        'artist_external_urls.spotify'], inplace=True)
    return album

```

```python
def format_artist(name):
    artist = unidecode(name)
    artist = artist.lower()
    artist = artist.replace(' o ', ' ')
    artist = artist.replace(' ', '-')
    artist = artist.replace('.', '')
    print(artist)
    
    if (artist == 'Bruno & Marrone') or (artist == 'Jorge & Mateus'):
        artist = artist.replace(' & ', '-e-')
    elif (artist == 'victor-&-leo'):
        artist = artist.replace('-&-', '-')
    else:
        artist = artist.replace('&', 'e')
    
    return artist
```

```python
def scrape_songs_URL(artist_name):
    songs_url = []
    r = requests.get(f'https://www.vagalume.com.br/{artist_name}/')
    print(r.url)
    soup = bs(r.text, 'html.parser')
    music_list = soup.find(id='alfabetMusicList')
    for i in music_list.find_all('a'):
        if '#play' not in i.get('href'):
            if '-traducao' not in i.get('href'):
                if 'traducao' not in i.get('href'):
                    songs_url.append(f'https://www.vagalume.com.br{i.get("href")}')
    return songs_url
```

```python
def scrape_lyrics(songs_url):
    r = requests.get(songs_url)
    soup = bs(r.content.decode('utf-8','ignore'))
    lyrics = soup.find(id='lyrics')
    for i in lyrics.find_all('br'):
        i.replace_with('\n')
    return lyrics.text
```

```python
def sent_to_words(sentences):
    '''
    Retirando os acentos e tokeniza a string.
    '''
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations
```

```python
data_words = list(sent_to_words(list_of_all_lyrics))
```

```python
# Constrói os modelos de bigramas e trigramas.
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=20) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=20)  

bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

### Estabelecendo a quantidade de tópicos

```python
def compute_coherence_values(corpus, texts, limit, start, step):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=num_topics, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=2000,
                                           passes=20,
                                           alpha='auto',
                                           per_word_topics=True)
        model_list.append(lda_model)
        coherencemodel = CoherenceModel(model=lda_model, texts=texts, coherence='u_mass')
        coherence_values.append(coherencemodel.get_coherence())
        print(f'Número de tópicos: {num_topics} - Valor de Coerência: {coherencemodel.get_coherence()}')

    return model_list, coherence_values
```

```python
# Parâmetros para testagem em massa do modelo
start=2
limit=20
step=1

model_list, coherence_values = compute_coherence_values(corpus=corpus, texts=data_lemmatized, start=start, limit=limit, step=step)
```

```python
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```
